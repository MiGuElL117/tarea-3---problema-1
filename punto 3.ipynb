```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np

# Aplicar un umbral a las predicciones para convertirlas en valores binarios
umbral = 5.5  # Umbral para considerar una calidad mayor a 5.5 como buena
y_pred_binario = np.where(y_pred > umbral, 1, 0)

# Calcular métricas de evaluación
accuracy = accuracy_score(y_test, y_pred_binario)
precision = precision_score(y_test, y_pred_binario, average='weighted', zero_division=1)  # Utiliza average='weighted'
recall = recall_score(y_test, y_pred_binario, average='weighted', zero_division=1)  # Utiliza average='weighted'
f1 = f1_score(y_test, y_pred_binario, average='weighted')  # Utiliza average='weighted'

print("Métricas de evaluación:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
```
